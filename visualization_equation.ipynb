## Step1.1: Visualization of random numbers.

### After generating 1000 random numbers, open and read "x_dataset.csv", then using these numbers to be y list. And then, count the total data and use these data to be x list and plot scatter graph to show the distribution of these random numbers.

```python
import csv
import matplotlib.pyplot as plt

count = 0
y = list()
x = list()

with open('x_dataset.csv', newline='') as x_readfile:
    reader_x = csv.reader(x_readfile)
    for row_x in reader_x:
        count += 1
        y.append(float(row_x[0]))
        x.append(count)
        
plt.figure(figsize=(8, 6))
plt.scatter(x, y, c='b', label='data')
plt.title('Visualization of random numbers', fontsize=10)
plt.xlabel("The number of data points", fontsize=10)
plt.ylabel("The values of random numbers", fontsize=10)
plt.show()
```

## Step 2.1: Visualization of equation y.

### Open and read "y_dataset.csv", then using these numbers to be y list. And then using the variable x that already created in step 1.1 and using these data to be x list and plot scatter graph to show the distribution of these y data of equation.

```python
equation_y = list()
total_x = list()
count = 0


with open('y_dataset.csv', newline='') as y_readfile:
    reader_y = csv.reader(y_readfile)
    for row_y in reader_y:
        count += 1
        equation_y.append(float(row_y[0]))
        total_x.append(count)
        
plt.figure(figsize=(8, 6))
plt.scatter(total_x, equation_y, c='b', label='data')
plt.title('Visualization of equation y', fontsize=10)
plt.xlabel("The number of data points", fontsize=10)
plt.ylabel("The values of y", fontsize=10)
plt.show()
```
